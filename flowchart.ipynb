```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
def draw_flowchart():
    fig, ax = plt.subplots(figsize=(8, 10))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 14)
    ax.axis("off")  # Hide axes

    # Function to draw boxes
    def draw_box(x, y, text, boxstyle="round", color="lightblue"):
        ax.text(x, y, text, ha="center", va="center", 
                bbox=dict(boxstyle=boxstyle, facecolor=color, edgecolor="black", linewidth=1))

    # Function to draw diamonds
    def draw_diamond(x, y, text, color="lightyellow"):
        diamond = patches.RegularPolygon((x, y), numVertices=4, radius=1, orientation=np.pi/4,
                                         edgecolor='black', facecolor=color, linewidth=1)
        ax.add_patch(diamond)
        ax.text(x, y, text, ha="center", va="center")

    # Function to draw arrows
    def draw_arrow(x1, y1, x2, y2):
        ax.annotate("", xy=(x2, y2), xytext=(x1, y1),
                    arrowprops=dict(arrowstyle="->", lw=1.5, color="black"))

    # Drawing Nodes
    draw_box(5, 13, "Start", "round", "lightgreen")
    draw_box(5, 11, "Initialize Grid")
    draw_diamond(5, 9, "Find Empty Cell", "lightyellow")
    draw_box(5, 7, "Try Numbers 1-9")
    draw_diamond(5, 5, "Is Number Safe?", "lightyellow")
    draw_box(3, 3, "Try Next Number", "round", "lightcoral")
    draw_box(7, 3, "Place Number")
    draw_box(7, 1, "Recurse to Next Cell")
    draw_box(5, -1, "Solution Found", "round", "lightgreen")

    # Drawing Arrows
    draw_arrow(5, 12.5, 5, 11.5)
    draw_arrow(5, 10.5, 5, 9.5)
    draw_arrow(5, 8.5, 5, 7.5)
    draw_arrow(5, 6.5, 5, 5.5)
    
    # Decision Box Arrows
    draw_arrow(5, 4.5, 3.5, 3.5)  # No -> Try Next Number
    draw_arrow(5, 4.5, 6.5, 3.5)  # Yes -> Place Number
    draw_arrow(7, 2.5, 7, 1.5)  # Recurse to Next Cell
    draw_arrow(3, 2.5, 5, -0.5)  # Try Next Number -> Solution Found
    draw_arrow(7, 0.5, 5, -0.5)  # Recurse to Next Cell -> Solution Found

    plt.show()

# Call the function to draw the flowchart
draw_flowchart()
```
